# Model Creation

```python
#Independent and Dependent variables creation

x=df[['age','income','spending_score','loyalty_years']]  #select features(independent variables) for the model
y=df.purchases                                           #select target variable(dependent variable) -Purchases
```

```python
x  #Displays the feature matrix containing columns: age,income,spending_score,loyalty_years
```

```python
y  #Displays the target variable(purchases) as a series
```

```python
#Training and Testing data creation

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train  #Displays the training features after train-test-split
```

```python
y_train #Displays the target values(purchases) for the training set after train-test-split
```

```python
#Creating Model

from sklearn.linear_model import LinearRegression  #imorting Linear Regression model from scikit-learn

lr=LinearRegression()  #Creating Instance(object) to the model
```

```python
lr.fit(x_train,y_train) #Training of Linear Regression
```

```python
y_pred=lr.predict(x_test)  #Predict purchases using the trained model on the test set features 
y_pred                     #Displays predicted purchases
```

```python
import warnings  #Suppress all warnings
warnings.filterwarnings("ignore")
```

```python
lr.predict([[30,60000,50,6]]) #Predicts purchases for a customer input: age(30),income(60000),spending_score(50),loyalty_years(6) 
```

y_test is the ground truth for the test set.
y_pred is the model's prediction for the sameÂ testÂ inputs.

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2_score=r2_score(y_test,y_pred)
r2_score
```

```python
from sklearn.metrics import mean_squared_error  # Import function to calculate Mean Squared Error (MSE) for model evaluation
mean_squared_error(y_test,y_pred)               # Calculates the Mean Squared Error between actual and predictedÂ salesÂ values
```

```python
mean_squared_error(y_test,y_pred)**0.5        # Calculates Root Mean Squared Error to measure average prediction error inÂ originalÂ units
```
